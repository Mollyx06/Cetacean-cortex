```python
## 添加特征基因保存成 H5AD
```

```python
genes=None
cell=None
mtx=None
for name in os.listdir(path1):
    file= os.path.join(path1, name)
    if 'barcodes.tsv' in name:
        cell = pd.read_csv(file, header=None)
    elif 'features.tsv' in name:
        genes = pd.read_csv(file, sep='\t', header=None)
    elif 'matrix.mtx' in name:
        mtx = mmread(file)

adata = anndata.AnnData(mtx)
adata=adata.T
adata.obs.index = pd.Index(cell[0])
adata.var.index = pd.Index(genes[0])
adata.var_names = genes[0].to_list()
adata.obs_names = cell[0].to_list()
adata.X = adata.X.A
adata.var_names_make_unique
adata.write('data/processed/raw_snRNA_h5ad_file')
```
