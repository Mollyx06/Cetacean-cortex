```r
data <- readRDS("data/processed/cortex_analysis_file")
data
```

```r
# 映射分组
meta <- data@meta.data
df_group <- meta[!duplicated(meta$annotation), c('major', 'annotation')]
group <- structure(df_group$major, names = df_group$annotation)
```

```r
x <- createNeuronChat(log_data, DB='human', meta=meta, group.by = meta$annotation)
x <- run_NeuronChat(x,M=100)
```

```r
# 保存整个NeuronChat对象
saveRDS(x, "results/neuronchat/neuronchat_output_file")
```
